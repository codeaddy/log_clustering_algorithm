```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer

# Выбор признаков для кластеризации
# features = ['case:RequestedAmount', 'NumberOfTerms', 'MonthlyCost', 'CreditScore', 'OfferedAmount']
# features = ['case:RequestedAmount', 'CreditScore']
# features = ['OfferedAmount', 'CreditScore'] # eps = 0.1, min_samples = 6
features = ['NumberOfTerms', 'CreditScore'] # eps = 0.07, min_samples = 5
# features = ['case:RequestedAmount', 'OfferedAmount', 'CreditScore'] # eps = 0.07, min_samples = 5 no...
# features = ['OfferedAmount', 'MonthlyCost', 'Process_Duration', 'case:RequestedAmount'] # eps = 0.1, min_samples = 6

data_subset = df[features]

# Заполнение пропущенных значений медианой
imputer = SimpleImputer(strategy='median')
data_filled = imputer.fit_transform(data_subset)

# Нормализация данных
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(data_filled)

# Применение DBSCAN
dbscan = DBSCAN(eps=0.07, min_samples=5)
clusters = dbscan.fit_predict(data_normalized)

# Добавление информации о кластерах
df['Cluster'] = clusters

# Сохранение результатов кластеризации
df.to_csv('output_dataset.csv', index=False)

unique_clusters = pd.unique(clusters)
print(f"Обнаружено кластеров: {len(unique_clusters)}")
print(df[['case:RequestedAmount', 'NumberOfTerms', 'MonthlyCost', 'CreditScore', 'OfferedAmount', 'Cluster']].head())
```

```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# numeric_features = ['case:RequestedAmount', 'NumberOfTerms', 'MonthlyCost', 'CreditScore', 'OfferedAmount']
numeric_features = ['Process_Duration']
categorical_features = ['Selected']
# df['concept:name'] = df['concept:name'].astype(str)
df['Selected'] = df['Selected'].astype(str)
# df['case:LoanGoal'] = df['case:LoanGoal'].astype(str)

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', MinMaxScaler())])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

dbscan = DBSCAN(eps=0.2, min_samples=20)

pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('clusterer', dbscan)])

df['Cluster'] = pipeline.fit_predict(df)

df.to_csv('output_dataset.csv', index=False)

unique_clusters = pd.unique(df['Cluster'])
print(f"Обнаружено кластеров: {len(unique_clusters)}")
print(df.head())
```

```python
import numpy as np

for eps in np.arange(0.1, 0.30, 0.03):
    for min_samples in range(2, 10):
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        clusters = dbscan.fit_predict(data_normalized)
        print(f"EPS: {eps}, Min Samples: {min_samples}, Clusters detected: {len(np.unique(clusters))}")
```

```python
eps_values = np.arange(0.2, 2.0, 0.2)
min_samples_values = range(10, 300, 10)

results = {}

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_transformer, numeric_features),
            ('cat', categorical_transformer, categorical_features)])

        pipeline = Pipeline(steps=[
            ('preprocessor', preprocessor),
            ('clusterer', dbscan)
        ])
        
        pipeline.fit(df)
        cluster_labels = pipeline.named_steps['clusterer'].labels_
        
        num_clusters = len(np.unique(cluster_labels)) - (1 if -1 in cluster_labels else 0)
        results[(eps, min_samples)] = num_clusters
        
        print(f"EPS: {eps}, Min Samples: {min_samples}, Clusters found: {num_clusters}")
```

```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

numeric_features = ['case:RequestedAmount', 'Process_Duration']
# categorical_features = ['Accepted']
df['Accepted'] = df['Accepted'].astype(str)

df['Process_Duration'] = df['Process_Duration'].dt.total_seconds() / 3600

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', MinMaxScaler())])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
#         ('cat', categorical_transformer, categorical_features)
    ])
```

```python
dbscan = DBSCAN(eps=0.125, min_samples=75)

pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('clusterer', dbscan)])

df['Cluster'] = pipeline.fit_predict(df)

df.to_csv('output_dataset.csv', index=False)

unique_clusters = pd.unique(df['Cluster'])
print(f"Обнаружено кластеров: {len(unique_clusters)}")
print(df[['CreditScore', 'Process_Duration', 'Accepted', 'Cluster']].head())
```

```python
eps_values = np.arange(0.08, 1.0, 0.005)
min_samples_values = range(10, 100, 5)

results = {}

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_transformer, numeric_features),
#             ('cat', categorical_transformer, categorical_features)
        ])

        pipeline = Pipeline(steps=[
            ('preprocessor', preprocessor),
            ('clusterer', dbscan)
        ])
        
        pipeline.fit(df)
        cluster_labels = pipeline.named_steps['clusterer'].labels_
        
        num_clusters = len(np.unique(cluster_labels)) - (1 if -1 in cluster_labels else 0)  # Исключаем шум (-1)
        results[(eps, min_samples)] = num_clusters
        
        # Вывод результатов
        print(f"EPS: {eps}, Min Samples: {min_samples}, Clusters found: {num_clusters}")
```

```python
import graphviz

# Пример процесса с переходами между событиями
edges = [
    ('Начало', 'A_Create Application'),
    ('A_Create Application', 'A_Submitted'),
    ('A_Submitted', 'W_Handle leads'),
    ('W_Handle leads', 'W_Complete application'),
    ('W_Complete application', 'Конец')
]

dot = graphviz.Digraph(comment='Пример процессной карты')

for edge in edges:
    dot.edge(edge[0], edge[1])

dot.render('process_map', view=True)  # Сохраняет и отображает карту процесса
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

data = df

G = nx.DiGraph()

for index, row in data.iterrows():
    if index < data.shape[0] - 1:
        source = row['concept:name']
        target = data.iloc[index + 1]['concept:name']
        G.add_edge(source, target)

plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=700, edge_color='k', linewidths=1, font_size=10, arrows=True)
plt.title('Сетевой граф BPI Challenge 2017')
plt.show()
```
